```python
data = pd.read_csv('countries.csv')
```

```python
data
```

```python
# Compare the population growth of India and China
```

```python
data [data.country == 'India']
```

```python
India = data [data.country == 'India']
```

```python
China = data [data.country == 'China']
```

```python
China
```

```python
plt.plot(India.year, India.population /10**6)
plt.plot(China.year, China.population /10**6)
plt.legend (['India', 'China'])
plt.xlabel('year')
plt.ylabel('population')
plt.show()
```

```python
India.population
```

```python
India.population / India.population.iloc[0] *100
```

```python
plt.plot(India.year, India.population / India.population.iloc[0] *100)
plt.plot(China.year, China.population / China.population.iloc[0] *100)
plt.legend (['India', 'China'])
plt.xlabel('year')
plt.ylabel('population growth (first year = 100)')
plt.show()
```
